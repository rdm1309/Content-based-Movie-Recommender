## Creating tags

##
[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}] 
### Converting the above Format(list of dictionaries) to ['Adventure','Fantasy','Science Fiction'] to create proper tags

```python
# Function to extract the value of name from every dictionary from the list of dictionary 'Genres'
def convert(list1):
    genreL=[]
    for i in ast.literal_eval(list1):
        genreL.append(i['name'])
    return genreL    
```

```python
Movie_dataFrame['genres']=Movie_dataFrame['genres'].apply(convert)
```

```python
Movie_dataFrame['genres']
```

```python
Movie_dataFrame['keywords']=Movie_dataFrame['keywords'].apply(convert)
```

```python
Movie_dataFrame['keywords']
```

```python
Movie_dataFrame
```

```python
# Function to extract the value of name from every dictionary from the list of dictionary 'Cast'
def convert(list1):
    castlist=[]
    count=0
    for i in ast.literal_eval(list1):
        if count!=3:
            castlist.append(i['name'])
            count+=1
        else:
            break
    return castlist       
```

```python
Movie_dataFrame['cast']=Movie_dataFrame['cast'].apply(convert)
```

```python
Movie_dataFrame['cast']
```

```python
Movie_dataFrame['crew'].values
```

```python
# FUNCTION TO EXTRACT DRIECTOR NAME FROM CREW COLUMN
def convert(list1):
        directorList=[]
        for i in ast.literal_eval(list1):
            if i['job']=='Director':
                directorList.append(i['name'])
                break
        return directorList
```

```python
Movie_dataFrame['crew']=Movie_dataFrame['crew'].apply(convert)
```

```python
Movie_dataFrame.head()
```

## Since all the columns are converted to list. We convert overview column that is a string to list to create tag(By combining some columns)

```python
Movie_dataFrame['overview']=Movie_dataFrame['overview'].apply(lambda x:x.split())
```

```python
Movie_dataFrame
```

```python
# USING LIST COMPREHENSION TO REMOVE SPACE BETWEEN NAMES FROM THE LIST 
Movie_dataFrame['genres']=Movie_dataFrame['genres'].apply(lambda x:[i.replace(" ","") for i in x])
Movie_dataFrame['cast']=Movie_dataFrame['cast'].apply(lambda x:[i.replace(" ","") for i in x])
Movie_dataFrame['keywords']=Movie_dataFrame['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
Movie_dataFrame['crew']=Movie_dataFrame['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
Movie_dataFrame
```

```python
#CREATING TAGS COLUMN USING GENRES, KEYWORDS, CAST, OVERVIEW, CREW COLUMN
Movie_dataFrame['tags']=Movie_dataFrame['overview']+Movie_dataFrame['genres']+Movie_dataFrame['keywords']+Movie_dataFrame['cast']+Movie_dataFrame['crew']
```

```python
Movie_dataFrame
```

```python
# CREATING NEW DATA FRAME WITH ID TITLE  TAGS COLUMNS
Final_dataFrame=Movie_dataFrame[['id','title','tags']]
```

```python
Final_dataFrame
```

```python
#converting the tags list to string for every row 
Final_dataFrame['tags']=Final_dataFrame['tags'].apply(lambda x:" ".join(x))
```

```python
Final_dataFrame
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#helper function
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
Final_dataFrame['tags']=Final_dataFrame['tags'].apply(stem)
```

```python
Final_dataFrame['tags'][0]
```

```python
#Optional
Final_dataFrame['tags']=Final_dataFrame['tags'].apply(lambda x:x.lower())
```

```python
Final_dataFrame['tags'][0]
```

```python
# STRING VECTORISATION
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(Final_dataFrame['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
len(cv.get_feature_names())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(Similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=Final_dataFrame[Final_dataFrame['title']==movie].index[0]
    distances=Similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:7]
    
    for i in movie_list:
        print(Final_dataFrame.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
